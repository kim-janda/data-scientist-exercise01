### Explore continuous values

```python
#univariate analysis of the continuous variables
cont = ['age', 'education_num', 'capital_gain', 'capital_loss', 'hours_week']

df[cont].describe()
```

Looking at the univariate analysis suggests that capital_loss and capital_gain might be mostly 0 for the majority of records. The lack of variability means they might not be as useful for a modeling exercise.

Next let's look at each variable against the response var `over_50k`

```python
#check the mean
print(df.groupby('over_50k')[cont].mean())

#median
print(df.groupby('over_50k')[cont].median())

#sd
print(df.groupby('over_50k')[cont].std())
```

The medians of both `capital_gains` and `capital_loss` by `over_50k` are 0 as well which suggests that there are a lot of zero values confirming they are unlikely to be useful in a modeling exercise.

`age`, `education_num`, and `hours_week` have differences in their means and std suggesting that the distributions
may be different enough to be informative to a model.
